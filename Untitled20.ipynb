```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Titanic.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Replacing 0 and 1 with 'Not Survived' and 'Survived'
df['Survived'] = df['Survived'].replace({0: 'Not Survived', 1: 'Survived'})

# Verify the change
df['Survived'].head()
```

```python
# checking the information
df.info()
```

```python
# Replace numerical values in 'Pclass' with descriptive labels
df['Pclass'] = df['Pclass'].replace({1: 'First Class', 2: 'Second Class', 3: 'Third Class'})

# Rename the column to 'Passenger Class'
df.rename(columns={'Pclass': 'Passenger Class'}, inplace=True)

# Verify the changes
df[['Passenger Class']].head()
```

```python
# Check unique values in 'Pclass' column
print(df['Passenger Class'].unique())
```

```python
df.info()
```

```python
df['Embarked'] = df['Embarked'].replace({
    'S': 'Southampton',
    'C': 'Cherbourg',
    'Q': 'Queenstown'
})
print(df['Embarked'].value_counts())
```

```python
df.head()
```

```python
df.rename(columns={'Parch': 'ParentsChildrenAboard'}, inplace=True)
df.rename(columns={'SibSp': 'SiblingsSpousesAboard'}, inplace=True)
df.columns
```

```python
df.rename(columns={'Passenger Class': 'PassengerClass'}, inplace=True)
df.columns
```

```python
df.head()
```

```python
df.info()
```

# We cleaned the dataset by renaming columns for better clarity. Survived was replaced with 'Not Survived' and 'Survived', and Pclass was changed to 'PassengerClass' with class names like 'First Class', 'Second Class', and 'Third Class'. We also renamed SibSp and Parch to SiblingsSpousesAboard and ParentsChildrenAboard, replaced port codes in Embarked with full names, and dropped the Cabin column due to excessive missing data.

```python
# Pie chart for 'Survived' column
survival_counts = df['Survived'].value_counts()

plt.figure(figsize=(7, 7))
survival_counts.plot.pie(autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Survived vs Not Survived')
plt.ylabel('')  # Removing the ylabel to keep it clean
plt.show()
```

# This will show the percentage of passengers who survived and those who didnt.

```python
# Histogram for 'Age' column
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'].dropna(), kde=True, bins=30, color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# This will show how the ages are distributed across the dataset.

```python
# Boxplot for 'Fare' column based on 'Survived'
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Fare', data=df, palette='Set2')
plt.title('Fare Distribution by Survival')
plt.show()
```

# This will help compare the fare distribution between passengers who survived and those who didnt.

```python
# Scatterplot for 'Age' vs 'Fare'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', data=df, hue='Survived', palette='coolwarm', alpha=0.7)
plt.title('Age vs Fare (Colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

# This will show if there's any noticeable trend or pattern between Age and Fare in terms of survival.

```python
# Ensure we're using only numerical columns for correlation
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate the correlation matrix
corr_matrix = df[numerical_cols].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1, linecolor='black')
plt.title('Correlation Heatmap')
plt.show()
```

# This will only include numerical columns in the correlation matrix and should eliminate the error.
